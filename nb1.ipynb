```python
from sklearn.model_selection import train_test_split 

X =  data.drop(columns= ['geohash_target','demand'] ,  axis = 1)
y =  data['demand']
```

```python
x_train,  x_val,y_train,y_val =  train_test_split(X,y,random_state= 42,test_size= 0.2)
```

```python
temp  =  x_train.copy()
temp['demand'] =  y_train 

geo_mean =  temp.groupby('geohash')['demand'].mean()
```

```python
x_train['geohash_target'] =  x_train['geohash'].map(geo_mean)
x_val['geohash_target'] =  x_val['geohash'].map(geo_mean)
```

```python
x_train.info()
```

```python
x_train
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error 
```

```python
cat_features = ['geohash']
```

```python
model =  CatBoostRegressor(
    iterations=5000,
    learning_rate=0.03,
    depth =8,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_seed= 42,
    early_stopping_rounds=200,
    verbose= 100
)

model.fit(x_train,
          y_train,
          eval_set=(x_val,y_val),
          cat_features= cat_features,
          use_best_model= True 
          )
```

```python
preds =  model.predict(x_val)
```

```python
rmse =  np.sqrt(mean_squared_error(y_val,preds))
```

```python
print("RMSE: ",rmse)
```
